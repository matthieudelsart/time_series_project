```python
import pandas as pd
```

```python
# load core data + holidays dataset
train_holidays = pd.read_csv("data/train_holidays.csv")
test_holidays = pd.read_csv("data/test_holidays.csv")

# load weather dataset
weather = pd.read_csv("data/weather_clean.csv")
weather.rename(columns={'date': 'id'}, inplace=True) # rename so that it matches the first dataset
```

```python
# merge the datasets

# Convert 'id' to datetime with format 'YYYY-MM-DD HH'
train_holidays['id'] = pd.to_datetime(train_holidays['id'], format='%Y-%m-%d %H')
test_holidays['id'] = pd.to_datetime(test_holidays['id'], format='%Y-%m-%d %H')

# Convert 'id' in weather to datetime
weather['id'] = pd.to_datetime(weather['id'])

# left join to create train with holidays + weather data
train = pd.merge(train_holidays, weather, on='id', how='left')
```

### Feature Engineering

Columns added:
- Seasons
- Weekday vs Weekend

```python
train['date'] = pd.to_datetime(train['id'])
train.set_index('date', inplace=True)

# 1. Season Column (Binary Columns for Each Season)
def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Fall'

# Map month to season
train['season'] = train.index.month.map(get_season)

# Convert season to binary columns
train['is_winter'] = (train['season'] == 'Winter').astype(int)
train['is_spring'] = (train['season'] == 'Spring').astype(int)
train['is_summer'] = (train['season'] == 'Summer').astype(int)
train['is_fall'] = (train['season'] == 'Fall').astype(int)
train.drop(columns=['season'], inplace=True)

# 2. Weekday vs. Weekend Column (Binary)
train['is_weekend'] = (train.index.dayofweek >= 5).astype(int)  # 1 for weekend, 0 for weekday
```
