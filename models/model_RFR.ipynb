```python
target_columns = ['valeur_NO2', 'valeur_CO', 'valeur_O3', 'valeur_PM10', 'valeur_PM25']
```

```python
X_train = df.drop(columns=['id'] + target_columns)  
y_train = df[target_columns] 
```

```python
X_test = test.drop(columns=['id'])
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_normalized = scaler.fit_transform(X_train)
X_test_normalized = scaler.transform(X_test) 
```

```python
# Convert normalized arrays to DataFrame for later use
X_train_normalized = pd.DataFrame(X_train_normalized, columns=X_train.columns)
X_test_normalized = pd.DataFrame(X_test_normalized, columns=X_test.columns)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.multioutput import MultiOutputRegressor

model = MultiOutputRegressor(RandomForestRegressor())
model.fit(X_train_normalized, y_train)  # train the model
```

```python
predictions = model.predict(X_test_normalized) 
```

```python
predictions_df = pd.DataFrame(predictions, columns=['valeur_NO2', 'valeur_CO', 'valeur_O3', 'valeur_PM10', 'valeur_PM25'])
```

```python
predictions_df['id'] = test['id'].values
```

```python
predictions_df = predictions_df[['id'] + target_columns]
```

```python
predictions_df = clean_for_submission(predictions_df)
```
